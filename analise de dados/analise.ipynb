# Verificação de outliers

```python
# Função para criar e exibir subplots com boxplots
def criar_subplots_boxplots(dados, colunas, titulos, limite_maximo):
    fig, axs = plt.subplots(1, len(colunas), figsize=(15, 5))
    for i, coluna in enumerate(colunas):
        axs[i].boxplot(dados[coluna].dropna())
        axs[i].set_title(titulos[i])
        axs[i].set_ylabel('Valores')
        axs[i].set_ylim(0, limite_maximo)
    plt.tight_layout()
    plt.show()

# Definindo colunas e títulos
colunas = ['NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_REDACAO']
titulos = [
    'Matemática',
    'Linguagens e Códigos',
    'Ciências da Natureza',
    'Ciências Humanas',
    'Redação'
]
```

## 2019

```python
print('ESCOLAS PÚBLICAS')
criar_subplots_boxplots(dados_publica_2019, colunas, titulos, limite_maximo=1000)

print('ESCOLAS PRIVADAS')
criar_subplots_boxplots(dados_privada_2019, colunas, titulos, limite_maximo=1000)

print('ESCOLAS URBANAS')
criar_subplots_boxplots(dados_urbana_2019, colunas, titulos, limite_maximo=1000)

print('ESCOLAS RURAIS')
criar_subplots_boxplots(dados_rural_2019, colunas, titulos, limite_maximo=1000)
```
